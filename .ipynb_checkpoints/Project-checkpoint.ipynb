```python
#create some simple functions for future use
def to_secs(time):
    '''
    Take a string representing a time and convert it to seconds
    
    input should be in M:S, ignore milliseconds
    output is an integer of seconds for ease of use later
    '''
    in_secs = 0
    time = time.split('.')[0]
    split_time = time.split(':')
    in_secs += int(split_time[0]) * 60
    in_secs += int(split_time[1])
    
    return in_secs
```
